```python
import cv2
```

```python
img_path = 'img_detect.png'
```

```python
img = cv2.imread(img_path)
```

```python
img.shape
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
# print(img)
```

```python
gray_img.shape
```

```python
face_classifier = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
)
```

```python
face = face_classifier.detectMultiScale(
    gray_img, scaleFactor=1.1, minNeighbors=5, minSize=(200,200)
)
```

```python
for (x, y, w, h) in face:
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 3)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
# flipping the img
flipped_img_hor = cv2.flip(img_rgb,1)
flipped_img_ver = cv2.flip(img_rgb,0)
flipped_img_both = cv2.flip(img_rgb,-1)
```

```python
mp.figure(figsize=[15,15])
mp.subplot(141);mp.imshow(flipped_img_hor);mp.title("Horizontal Flip")
mp.subplot(142);mp.imshow(flipped_img_ver);mp.title("Vertical Flip")
mp.subplot(143);mp.imshow(flipped_img_both);mp.title("Horizontal & Vertical Flip")
mp.subplot(144);mp.imshow(img_rgb);mp.title("Original")
```

```python
line_img = img_rgb.copy()
cv2.line(line_img, (180,200),(210,200),(255,0,0),3,lineType=cv2.LINE_AA)
mp.imshow(line_img)
```

```python
circle_img = img_rgb.copy()
cv2.circle(circle_img,(187,310),26,(255,255,0),3)
mp.imshow(circle_img)
# mp.imshow(circle_img[:,:,::-1])
```
