```python
import cv2
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("New_Zealand_Coast.jpeg")
# print(img)
```

```python
img_bgr = cv2.imread("New_Zealand_Coast.jpeg", cv2.IMREAD_COLOR)
# plt.imshow(img_bgr)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
# plt.imshow(img_rgb)
```

```python
matrix = np.ones(img_rgb.shape, dtype="uint8") * 50

brighter_img = cv2.add(img_rgb, matrix)
darker_img = cv2.subtract(img_rgb, matrix)
```

```python
plt.figure(figsize=(20, 10))
plt.subplot(131);plt.imshow(brighter_img);plt.title("Brighter")
plt.subplot(132);plt.imshow(img_bgr);plt.title("Original")
plt.subplot(133);plt.imshow(darker_img);plt.title("Darker")
```
